```python
import json
import pandas as pd
import os

tweets_filename = os.getcwd() + '\Library Project Files\\newresults.txt'
tweets_file = open(tweets_filename, 'r')
data = []
```

```python
for line in tweets_file:
    try:
        tweet = json.loads(line.strip())
        if 'text' in tweet:
            textweet = tweet['text']
            if 'RT' not in textweet:
                mined = {'username':tweet['user']['name'], 'tweet':tweet['text']}
                data.append(mined)                    
           
           
    except:
        continue
```

```python
print(len(data))
mydf = pd.DataFrame(data)
mydf
```
